# Coding the Game

<p style='font-size:1.75rem;line-height:1.5'>
    Now that we have completed our game play drawing, let's create it <b style="color:red">using functions!</b>   
    </p>
 
<p style='font-size:1.75rem;line-height:1.5'>
    Things to remember: 
    <ul style='font-size:1.75rem;line-height:2'>
        <li><code>raw_input()</code> <b style="color:blue">returns a string</b> (be careful if you are asking the user for numbers!)</li>
        <li><code>'\n'</code> will help with <b style="color:blue">formatting the game</b></li>
        <li> For <b style="color:blue">each room</b>, we need to:
            <ol>
                <li> explain the <b>current situation</b> in the room</li>
                <li> explain the <b>choices</b></li>
                <li> ask for <b>user input</b></li>
                <li> <b>update</b> current situation based on user input, and then decide which room to go to next</li>
            </ol>  
        </li>
    </ul>
</p>

    We have created a sample <b style="color:red">STARTING ROOM</b> and <b style="color:blue">REPLAY ROOM</b> for you below. 
    <br> Use these to get you started, and add our own rooms!

```python
# starting room 
def starting_room():
    # ask user for their name 
    name = raw_input("What is your name? ")
    
    # explain the situation 
    print("\nHello, " + name + ".")
    print("Explain the situation in the Starting Room.")
    print("Choices: 1. Stay in Room 2. End Game")
   
    #get user choice 
    choice = raw_input("Enter a number: ")
          
    # check user choices 
    if choice == '1':
        print('\nYou chose to stay in the room')
        starting_room()
    elif choice == '2':
        print('\nYou chose to end the game')
        replay()
    else:
        print('\nYou entered an invalid input. We will re-enter the room. ')
        starting_room ()

##########################
# Add your functions/rooms here








##########################

def replay():
    #get user choice 
    choice = raw_input('Do you want to play again? Enter "y" or "n" (without quotes): ')
          
    # check user choices 
    if choice == 'y':
        starting_room()
    elif choice == 'n':
        end_of_game()
    else:
        print('\nPlease enter a valid input.')
        replay()
        
def end_of_game():
    print("\nThanks for playing!")

starting_room()

```
